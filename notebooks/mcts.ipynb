```python
import sys
import os

# Ensure repo root is on path
repo_root = "/home/union/quant/RLC2026"
if repo_root not in sys.path:
    sys.path.insert(0, repo_root)

from lib.grid import competitive_fourrooms_env
from src.planning import MCTSPlanner

env = competitive_fourrooms_env(render_mode="human")
env.reset(seed=0)
```

```python
planner = MCTSPlanner(
    action_space=env.action_space(env.possible_agents[0]),
    n_simulations=200,
    max_depth=50,
    gamma=0.99,
)
```

```python
for _ in range(100):
    agent = env.agent_selection
    action = planner.plan(env)  # uses deepcopy(env)
    env.step(action)

    env.render()
    

    if any(env.terminations.values()) or any(env.truncations.values()):
        break

env.close()
```
